> 他们相同的点就在于说明符是一样的，我们这里先来认识一下所有的说明符

# 1. 使用 .format() 方法格式化字符串

```python
# 对齐和宽度
print("{:<10}".format("Left"))      # 左对齐
print("{:>10}".format("Right"))     # 右对齐
print("{:^10}".format("Center"))    # 居中对齐
print("{:10}".format("Default"))    # 默认为左对齐
print("{:=^10}".format("Center"))   # 用 '=' 填充并居中对齐
```

```python
# 数字格式化
print("{:.2f}".format(3.1415926))   # 保留两位小数
print("{:,}".format(1000000))       # 千位分隔符
print("{:+}".format(15))            # 显示正号
print("{:-}".format(-15))           # 显示负号
print("{:05}".format(7))            # 用零填充，宽度为 5
```

```python
# 字符串格式化
print("{:.5}".format("Hello, World!"))  # 字符串截断为最大长度为 5
```

```python
# 类型转换
print("{:b}".format(10))    # 二进制表示
print("{:o}".format(10))    # 八进制表示
print("{:x}".format(10))    # 小写十六进制表示
print("{:X}".format(10))    # 大写十六进制表示
print("{:c}".format(65))    # Unicode 字符表示
```

```python
# 参数编号和关键字
print("{0} {1}".format("John", "Doe"))        # 位置编号
print("{first} {last}".format(first="John", last="Doe"))  # 关键字
```

# 2. 使用 f-string 实现上述 format() 的所有操作

```python
# 对齐和宽度
print(f"{'Left':<10}")             # 左对齐
print(f"{'Right':>10}")            # 右对齐
print(f"{'Center':^10}")           # 居中对齐
print(f"{'Default':10}")           # 默认为左对齐
print(f"{'Center':=^10}")          # 用 '=' 填充并居中对齐
```

```python
# 数字格式化
print(f"{3.1415926:.2f}")          # 保留两位小数
print(f"{1000000:,}")              # 千位分隔符
print(f"{+15:+}")                  # 显示正号
print(f"{-15:-}")                  # 显示负号
print(f"{7:05}")                   # 用零填充，宽度为 5
```

```python
# 字符串格式化
print(f"{'Hello, World!':.5}")     # 字符串截断为最大长度为 5
```

```python
# 类型转换
print(f"{10:b}")                   # 二进制表示
print(f"{10:o}")                   # 八进制表示
print(f"{10:x}")                   # 小写十六进制表示
print(f"{10:X}")                   # 大写十六进制表示
print(f"{65:c}")                   # Unicode 字符表示
print(f"{1000:,}")                 # 数字分隔符
```
