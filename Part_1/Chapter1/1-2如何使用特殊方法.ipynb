# 特殊方法
---
> 我们已经知道了，双下划线方法、魔术方法就是特殊方法<br>
> 这种方法常常适用于那些 python 内置方法的重构上面<br>
> 譬如说自己定义了一个类，但是对于某些数据的操作 **（in len() for ...）**，无法通过操作类的对象实现<br>
> 我们就通过对于基本方法重载，实现对象的操作

# 1.2.1 模拟数字类型
> 实现一个向量运算，我们创建向量的对象，然后实现一下运算操作<br>
> 1. 加法
> 2. 取模
> 3. 乘法
> 4. 打印时，输出自定义的Vector

**为此我们需要重载这几个特殊方法: \__repr__、\__abs__、 \__add__ 和 \__mul__**

一下的代码，不运行就是拆分开来做解析，运行看 testFile 文件下


```python
from math import hypot

class Vector:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def __repr__(self):
        """
        print() 向量的时候，输出自定义类，按照这个格式输出
        """
        return f'Vector({self.x}, {self.y})'

    def __abs__(self):
        """
        返回向量的模：sqrt(x**2, y**2)
        """
        return hypot(self.x, self.y)

    def __bool__(self):
        """
        如果是 False 就是没有方向的 0向量
        """
        return bool(abs(self))

    def __add__(self, other):
        """
        实现向量之间的 + 重载
        """
        x = self.x + other.x
        y = self.y + other.y
        return Vector(x, y)

    def __mul__(self, scalar):
        """
        实现向量的乘法重载
        """
        return Vector(self.x * scalar, self.y * scalar)

v1 = Vector(1,2)

v1
```
