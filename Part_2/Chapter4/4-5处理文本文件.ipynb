### 示例 4-10 细节分析上面的问题

```python
fp = open('./data/demo2.txt', 'w', encoding='utf-8')
fp
```

```python
fp.write('café')
fp.close()
```

```python
import os

os.stat('./data/demo2.txt').st_size
```

```python
fp2 = open('./data/demo2.txt')
fp2     # 就是这个对象，关注一下 encoding 的模式
```

```python
fp2.encoding
```

```python
# 使用指定的方法打开
fp3 = open('./data/demo2.txt', encoding='cp437')

fp3.encoding
```

```python
# 二进制模式打开
fp4 = open('./data/demo2.txt', 'rb')
print(fp4.read())
fp4.close()
```
