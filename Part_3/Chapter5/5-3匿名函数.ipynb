# 5.3 匿名函数
---
> lambda 关键字创建匿名函数，lambda函数的定义体中不能赋值 <br>
> 也不能使用 while 和 try 等 python 语句

## 示例 5-7 使用 lambda 实现反转字符串排序

```python
fruits = ['strawberry', 'fig', 'apple', 'cherry', 'raspberry', 'banana']

sorted(fruits, key= lambda word: word[::-1])
```

```python
is_even: str = lambda num: "Even" if num % 2 == 0 else "Odd"

is_even(20)
```
